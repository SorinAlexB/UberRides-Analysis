```python
#for dataset
import pandas as pd 

#visualisation
import matplotlib.pyplot as plt

#read the dataset
uber_rides= pd.read_csv("uber-raw-data-sep14.csv")
```

```python
#changing the "Date/Time" column's data type from string to datetime
uber_rides['Date/Time']= pd.to_datetime(uber_rides['Date/Time'])

#convert "Date/Time" column elements from string to datetime
uber_rides["Day"] = uber_rides["Date/Time"].apply(lambda x: x.day)
uber_rides["Hour"] = uber_rides["Date/Time"].apply(lambda x: x.hour)
uber_rides["Weekday"] = uber_rides["Date/Time"].apply(lambda x: x.weekday())
```

```python
#visualize the density of rides per day of month
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(uber_rides.Day, width= 0.6, bins= 30)
plt.title("Density of trips per Day", fontsize=16)
plt.xlabel("Day", fontsize=16)
plt.ylabel("Density of rides", fontsize=16)
```

```python
#visualize the density of rides per weekday
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(uber_rides.Weekday, width= 0.6, range= (0, 6.5), bins=7, color= "green")
plt.title("Density of trips per Weekday", fontsize=16)
plt.xlabel("Weekday", fontsize=14)
plt.ylabel("Density of rides", fontsize=14)
```

```python
#visualize the density of rides per hour
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(uber_rides.Hour, width= 0.6, bins=24, color= "orange")
plt.title("Density of trips per Hour", fontsize=16)
plt.xlabel("Hour", fontsize=14)
plt.ylabel("Density of rides", fontsize=14)
```

```python
#visualize the Density of rides per location
fig,ax = plt.subplots(figsize = (12,6))
x= uber_rides.Lon
y= uber_rides.Lat
plt.scatter(x, y, color= "purple")
plt.title("Density of trips per Hour", fontsize=16)
plt.xlabel("Hour", fontsize=14)
plt.ylabel("Density of rides", fontsize=14)
```
